```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Data Set.csv", sep=';')
df.head()
```

```python
# to delete any coloum
df = df.drop(columns=['Date'])
# altarnative way: df.drop('Date', axis=1)
```

```python
# to see or read data set
df.head()
```

```python
plt.scatter(df['Starting (USD)'], df['Ending (USD)'])
plt.xlabel('Starting (USD)')
plt.ylabel('Ending (USD)')
plt.title('Stock Prices')
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df['Starting (USD)']
y = df['Ending (USD)']
```

```python
x.head()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(), color='red')
plt.scatter(df['Starting (USD)'], df['Ending (USD)'])
plt.xlabel('Starting (USD)')
plt.ylabel('Ending (USD)')
plt.title('Stock Prices')
```

```python
# to train the model
model = LinearRegression()
```

```python
model.fit(df[['Starting (USD)']],df[['Ending (USD)']])
# model.fit(x, y ) not working
```

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
# mx + C
m*16700 + c
```

```python
# to predict
model.predict([[16700]])
```

```python
df['Predicted_Y'] = model.predict(df[['Starting (USD)']])
df.head()
```

```python
plt.plot(df[['Starting (USD)']], df[['Predicted_Y']])
```

```python
plt.plot(df[['Starting (USD)']], df[['Predicted_Y']])
plt.scatter(x.mean(),y.mean(), color='red')
plt.scatter(df['Starting (USD)'], df['Ending (USD)'])
plt.xlabel('Starting (USD)')
plt.ylabel('Ending (USD)')
plt.title('Stock Prices')
```

```python
# 1st import all necessary library
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# upload working file or data set
df = pd.read_csv('/content/drive/MyDrive/Data Set for ML/Data Set.csv', sep=';')
df
```

```python
# then trean the model with any algorithm
model = LinearRegression()
model.fit(df[['Starting (USD)']],df[['Ending (USD)']])
```

```python
# declear feature
x = df[['Starting (USD)']]
y = df[['Ending (USD)']]

# declear all const value
m = model.coef_
m
c = model.intercept_
c
df['Predicted_Y'] = model.predict(df[['Starting (USD)']])
df.head()
```

```python
plt.plot(df[['Starting (USD)']], df[['Predicted_Y']])
plt.scatter(x.mean(),y.mean(), color='red')
plt.scatter(df['Starting (USD)'], df['Ending (USD)'])
plt.xlabel('Starting (USD)')
plt.ylabel('Ending (USD)')
plt.title('Stock Prices')
```

```python
df['Lose'] = df['Ending (USD)'] - df['Predicted_Y']
df.head()
```

```python
# MSE
from sklearn.metrics import mean_squared_error
# for MAE
# from sklearn.metrics import mean_absolute_error

mse = mean_squared_error(df['Ending (USD)'], df['Predicted_Y'])
mse
```

**Accuracy for Regression**
